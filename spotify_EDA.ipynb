```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
saved_tracks = pd.read_csv('my_saved_tracks.csv')
```

```python
saved_tracks.head()
```

```python
saved_tracks.rename({'Unnamed: 0': 'recency'}, axis='columns', inplace = True)
```

```python
saved_tracks.describe()
```

```python
saved_tracks.info()
```

```python
# Factorize some variables:

saved_tracks['key_factor'] = pd.factorize(saved_tracks['key'])[0]
saved_tracks['mode_factor'] = pd.factorize(saved_tracks['mode'])[0]
saved_tracks['ts_factor'] = pd.factorize(saved_tracks['time_signature'])[0]
```

```python
saved_tracks[['key_factor', 'mode_factor', 'ts_factor']].hist()
plt.tight_layout()
plt.show()
```

```python
# Pull out the numerical features:
interesting_audio_features = saved_tracks[['popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
```

```python
# Histograms of numerical features:
interesting_audio_features.hist(figsize = (7,8))
plt.tight_layout()
plt.show()
```

```python
# How are the different audio features related to one another?
corr_matrix = interesting_audio_features.corr()

palette = sns.diverging_palette(20, 220, n=256)

ax = plt.axes()
sns.heatmap(corr_matrix, cmap = palette, annot=True, fmt = '.2f', ax = ax)
ax.set_title('R^2 Coefficient Among Audio Features in Joseph\'s Spotify Library \n')
plt.show()
```

```python
# Has my choice in music become more or less popular over time?

ax = plt.axes()

sns.regplot(x = 'recency', y = 'popularity', data = saved_tracks, ax = ax, lowess = True).set(title = 'Popularity vs. Recency of Songs in my Library')
ax.invert_xaxis()
ax.set_xlabel('Relative Age in my Library')

plt.show()

# Check for correlation:
saved_tracks[['recency', 'popularity']].corr().iloc[0,1]
```
